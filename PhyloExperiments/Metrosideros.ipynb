```python
import pandas as pd
```

```python
#Read in species dataframe ##Change this to remove select only specific columns from original downloaded occurence sheet
metrosideros = pd.read_csv("occ.csv",index_col="genus")
metrosideros
```

```python
#Remove all entries with subspecies data to subset the original dataframe
metrosideros['infraspecificEpithet'] = metrosideros['infraspecificEpithet'].fillna('none')
metrosideros_og = metrosideros[metrosideros['infraspecificEpithet'] != 'none']
display(metrosideros_og)
```

```python
#Disable error for SettingWithCopyWarning before merging column contents
import pandas as pd
pd.options.mode.chained_assignment = None

#Merge species/subsp. names
metrosideros_og["species"] = metrosideros_og["specificEpithet"] + ' ' + metrosideros_og["infraspecificEpithet"]
metrosideros_og
```

```python
#delete individual species/subspecies columns
metrosideros_og = metrosideros_og.drop(['specificEpithet','infraspecificEpithet'],axis=1)
metrosideros_og
```

```python
#Remove all entries withOUT subspecies data to subset the original dataframe
metrosideros_nosub = metrosideros[~metrosideros.infraspecificEpithet.str.contains("var")]

#change column name to match new species column in the subspecies table
metrosideros_nosub.rename(columns={'specificEpithet':'species'}, inplace=True)
metrosideros_nosub
```

```python
#delete extraneous column with 'none'
metrosideros_nosub = metrosideros_nosub.drop(['infraspecificEpithet'],axis=1)
metrosideros_nosub
```

```python
#concatenate the 
NewTable = pd.concat([metrosideros_nosub, metrosideros_og])
NewTable
```

```python
LatTable = NewTable.groupby(['species']).agg({'decimalLatitude': ['min', 'max']})
LatTable
```

```python
LonTable = NewTable.groupby('species').agg({'decimalLongitude': ['min', 'max']})
LonTable
```

```python
#merge the aggragated dataframes by species
combined = pd.concat([LatTable, LonTable]).groupby('species').sum()
combined.insert(0, 'Genus', 'Metrosideros')
combined
```

```python
combined.reset_index(inplace=True)
combined.set_index('Genus',inplace=True)
combined
```
