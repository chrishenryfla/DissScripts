```python
#Create a shape

#Fill in latitude and longiutde coordinates, put 'points' in order to create the proper shape
lat_point_list = [combined['decimalLatitude']['min'][1], 
                  combined['decimalLatitude']['min'][1],
                  combined['decimalLatitude']['max'][1],
                  combined['decimalLatitude']['max'][1],
                  combined['decimalLatitude']['min'][1]
                 ]
lon_point_list = [combined['decimalLongitude']['min'][1],
                  combined['decimalLongitude']['max'][1],
                  combined['decimalLongitude']['max'][1],
                  combined['decimalLongitude']['min'][1],
                  combined['decimalLongitude']['min'][1]
                 ]
#Create polygon
polygon_geom = Polygon(zip(lon_point_list, lat_point_list))
crs = 'epsg:4326'
macropus = gpd.GeoDataFrame(index=[0], crs=crs, geometry=[polygon_geom])
```

```python
#Create shape files
macropus.to_file(filename='macropus.geojson', driver='GeoJSON')
macropus.to_file(filename='macropus.shp', driver="ESRI Shapefile")
```

```python
#Now Lets create a loop to create shapes for the rest of the species

speciestype = combined['species']

#Create polygon
metrosideros = {} #define the dictionary

for i in range(0,len(speciestype)):
    lat_point_list = [combined['decimalLatitude']['min'][i], 
                  combined['decimalLatitude']['min'][i],
                  combined['decimalLatitude']['max'][i],
                  combined['decimalLatitude']['max'][i],
                  combined['decimalLatitude']['min'][i]
                 ]
    lon_point_list = [combined['decimalLongitude']['min'][i],
                  combined['decimalLongitude']['max'][i],
                  combined['decimalLongitude']['max'][i],
                  combined['decimalLongitude']['min'][i],
                  combined['decimalLongitude']['min'][i]
                 ]
    
    metrosideros[i] = Polygon(zip(lon_point_list, lat_point_list))
    crs = 'epsg:4326'
    metrosideros[i] = gpd.GeoDataFrame(index=[0], crs=crs, geometry=[Polygon(zip(lon_point_list, lat_point_list))])
```

```python
x = list(metrosideros.keys())
for i in range(0,len(x)):
    metrosideros[speciestype[i]] = metrosideros.pop(x[i])
```

```python
#Create geo and shape files for each species
for i in metrosideros.keys():
    temp_filename = "%s.geojson" % i
    metrosideros[i].to_file(filename=temp_filename, driver='GeoJSON')
```

```python
for i in metrosideros.keys():
    temp_filename = "%s.shp" % i
    metrosideros[i].to_file(filename=temp_filename, driver="ESRI Shapefile")
```
